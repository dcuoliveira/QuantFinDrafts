```python
import pandas as pd
import numpy as np
import os
import pickle
import matplotlib.pyplot as plt
import seaborn as sb
import plotly.graph_objects as go
import plotly.io as pio

pio.renderers.default = 'notebook' # or 'notebook' or 'colab' or 'jupyterlab'
current_dir = os.getcwd()
```

```python
file = open(os.path.join(current_dir, "data", "signals_fast_fxmm.pickle"), 'rb')
signals = pickle.load(file)['fast']

instruments = list(signals.keys())
tags = {'USDBRL': 'BRL', 'USDCLP': 'CLP', 'USDCOP': 'COP', 'USDZAR': 'ZAR', 'USDMXN': 'MXN',
        'USDHUF': 'HUF', 'USDPLN': 'PLN', 'USDCZK': 'CZK', 'USDKRW': 'KRW', 'USDJPY': 'JPY',
        'USDAUD': 'AUD', 'USDCAD': 'CAD', 'USDNZD': 'NZD', 'USDGBP': 'GBP', 'USDNOK': 'NOK',
        'USDSEK': 'SEK', 'USDEUR': 'EUR', 'USDCHF': 'CHF'}

tags_inverted = {}
for k, v in tags.items():
    tags_inverted[v] = k
```

```python
def plot_ts_heatmap(df,
                    color_map="coolwarm",
                    title='Ranking dos sinais agregados para cada moeda \n\n\n',
                    annot=False):

    # heatmap
    fig, ax = plt.subplots(figsize=(20, 10))# plot heatmap
    sb.heatmap(df,
               cmap=color_map,
               vmin=df.min().min(),
               vmax=df.max().max(),
               square=True,
               linewidth=0.3,
               cbar_kws={"shrink": .8},
               annot=annot)
    ax.xaxis.tick_top()
    xticks_labels = df.columns
    plt.xticks(np.arange(len(df.columns)) + .6, labels=xticks_labels)# axis labels
    plt.xlabel('')
    plt.ylabel('')
    title = title.upper()
    plt.title(title, loc='left')
```

# Aggregate signals for each instrument

## 1. Largest weekly changes

```python
agg_signals = []
for ticker in instruments:
    agg_signals.append(signals[ticker].mean(axis=1))
    
agg_signals_df = pd.concat(agg_signals, axis=1)
agg_signals_df.columns = instruments

weekly_signals_df = agg_signals_df.resample("W").last() * -1

# padroniza os dados
weekly_signals_df.index = [dtref.strftime("%Y-%m-%d") for dtref in weekly_signals_df.index]

if tags is not None:
    for col in weekly_signals_df.columns:
        weekly_signals_df.rename(columns={col: tags[col]}, inplace=True)
```

### (a) Rankings

```python
plot_ts_heatmap(df=(weekly_signals_df.iloc[-4:]).rank(axis=1, ascending=False),
                color_map="coolwarm",
                title='Ranking dos sinais agregados para cada moeda \n\n\n')
```

### (b) Levels

```python
plot_ts_heatmap(df=weekly_signals_df.iloc[-4:],
                color_map="coolwarm_r",
                title='Nvel dos sinais agregados para cada moeda \n\n\n',
                annot=True)
```

## 2. Best/Worse currencies decomposition

```python
rank_df = weekly_signals_df.rank(axis=1, ascending=False)
best_worst_intruments = list(rank_df.iloc[-1].nlargest(2).index) + list(rank_df.iloc[-1].nsmallest(2).index)
ticker = best_worst_intruments[0]
```

```python
rank_df.iloc[-1].nlargest(2)
```

```python
rank_df.iloc[-1].nsmallest(2)
```

```python
weekly_ticker_signals_df = signals[tags_inverted[ticker]].resample("W").last() * -1
```

```python
categories = weekly_ticker_signals_df.columns
positive_values = weekly_ticker_signals_df[weekly_ticker_signals_df > 0].iloc[-1].fillna(0)
negative_values = weekly_ticker_signals_df[weekly_ticker_signals_df < 0].iloc[-1].fillna(0)

fig = go.Figure()
fig.add_trace(go.Scatterpolar(
      r=positive_values,
      theta=categories,
      fill='toself',
      name=''
))
fig.add_trace(go.Scatterpolar(
      r=negative_values,
      theta=categories,
      fill='toself',
      name=''
))

fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[np.min([positive_values.min(), negative_values.abs().min()]), np.max([positive_values.max(), negative_values.abs().max()])]
    )),
  showlegend=False
)

fig.show()
```
